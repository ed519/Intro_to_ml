```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Օգտագործեք pandas module-ը "car_data.csv" dataset-ը  import անելու համար և ստեղծեք pandas dataframe։

```python
df = pd.read_csv('car_data.csv')
```

### Տպեք dataframe-ի առաջին 5 արժեքները։

```python
df.head()
```

### Որո՞նք են 'make' սյան unique արժեքները։

```python
df['make'].unique()
```

### Ո՞ր 'make'-ն ունի ամենամեծ միջին գինը։ 

```python
df[df['price'] == df['price'].max()]['make']
```

### Ստեղծեք նոր dataframe, որը կպարունակի միայն 'wheel_base', 'length', 'width', 'height', 'curb_weight', 'horsepower', 'price' columns-ները։

```python
df1 = df[['wheel_base','length','width','height','curb_weight','horsepower','price']]
df1
```

### Օգտագործեք correlation matrix ու որոշեք թե որ feature-ներն են կորելացված 'price' feature-ի հետ։

```python
df1.corr()
```

### Գծեք scatter plot-եր (յուրաքանչյուր feature VS price)։ Ընտրեք 1 feature, որն ունի ամենագծային կախվածությունը 'price' feature-ից։

```python
X = df1[df1.columns.difference(['price'])]
y = df1['price']
fig, axis = plt.subplots(6, 1)
fig.set_figwidth(10)
fig.set_figheight(20)
fig.tight_layout(h_pad = 5)

for i, feature in enumerate(X.columns.tolist()):
    axis[i].scatter(df1[feature], y)
    axis[i].set_xlabel(feature)
    axis[i].set_ylabel('Price')
```

### Մենք կօգտագործենք 'price'-ը որպես target variable, ինչպես նաև Ձեր ընտրած feature-ը ու կկառուցենք linear regression model.

### Dataframe-ից վերցրեք միայն Ձեր ընտրած feature-ն ու 'price' feature-ը։ Ստուգեք թե արդյոք dataframe-ը պարունակում է Nan արժեքներ և լցրեք դրանք (fill) Ձեր ցանկացած մեթոդով։

```python
df1['horsepower'].isna().sum(), df1['price'].isna().sum()
```

```python
df1['horsepower'] = df1.fillna(method='bfill')
df1['price'] = df1.fillna(method='bfill')
```

```python
df1['horsepower'].isna().sum() , df1['price'].isna().sum()
```

```python
df1.corr()
```

### Բաժանեք տվյալները train և test dataset-երի 80:20 հարաբերակցությամբ։

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(df1['curb_weight'], df1['price'], test_size=0.2, random_state=50)
```

### Import արեք համապատասխան գրադարաններ fit արեք linear regression մոդել մեր տվյալներին։ Prediction կատարեք test տվյալների վրա։

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
#X_train = np.array().reshape(-1,1)
model.fit(X_train.values.reshape(-1,1),y_train)

model.coef_ , model.intercept_
```

### Import արեք համապատասխան library-ներ ու հաշվեք root mean squared error (RMSE) test տվյալների վրա։

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = model.predict(X_test.values.reshape(-1,1))

print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)),'\n'"")
print('R2 score:', r2_score(y_test,y_pred))
```

### Ուսումնասիրեք 'price' սյան տվյալներն ու որոշեք (ասեք) թե որքանով լավ արդյունք եք ստացել նախորդ առաջադրանքում։

```python
df1.corr()
```

```python
plt.scatter(df1['curb_weight'],df1['price'])
plt.scatter(X_test,y_pred , c='r')
plt.show()
```

### (Extra) Փորձեք օգտագործել ավելի շատ feature-ներ, արդյո՞ք ավելի լավ արդյունք եք ստանում։

```python
X_train, X_test, y_train, y_test = train_test_split(df1[df1.columns.difference(['price'])], df1['price'], test_size=0.2, random_state=50)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)),'\n'"")
print('MAE : ', mean_absolute_error(y_test, y_pred),'\n')
print('R2 score:', r2_score(y_test,y_pred))
```
