### Delete the columns 'Unnamed: 0' and 'Id'

### Ջնջեք 'Unnamed: 0' ու 'Id' սյուները

```python
baby_name = baby_name.drop(['Unnamed: 0', 'Id'], axis=1)
baby_name
```

### Are there more male or female names in the dataset?

### Արդյոք dataset-ում ավելի շատ են կանացի թե տղամարդու անունները?

```python
m = baby_name[baby_name['Gender'] == 'M']['Gender'].count()
f = baby_name[baby_name['Gender'] == 'F']['Gender'].count()
print("F",f)
print("M",m)
if m > f:
    print("M>F")
elif m<f:
    print("F<M")
elif f ==m:
    print("F=M")
```
